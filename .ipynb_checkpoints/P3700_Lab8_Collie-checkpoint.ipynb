***1. Histogram Equalization.*** *In the worksheet earlier this week, you adjusted contrast and brightness by hand. An automated approach to this is histogram equalization which uses the image’s own histogram to re-weight the gray scale levels in the image.*

$$ Gnew = \frac{\#Pixels with gray scale level ≤ G}{Total pixels}*(Max gray scale level)
$$

*(a) Write a function that takes an image as a 2D list and returns a 2D list of the image after histogram equalization has been applied.*

```python
import scipy.misc
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import time
import matplotlib.image as mpimg
ascent = scipy.misc.ascent()
```

```python
# This function takes the image and uses the Gnew calculation above to do adjust do an auto
#  adjust on the image. 

def hist_equal(image_2D):
    image_1D = image_2D.ravel()    # covert image to a 1D image
    max_grey = max(image_1D)       # find the max grey level in the image
    total_pixels = len(image_1D)  
    image_1D_equal = []
    
    # counts for each grey scale pixel
    grey_counts, binEdges = np.histogram(image_1D, bins=256)
    
    # adjust the pixel according the the Gnew calc aboove
    for n in image_1D:
        image_1D_equal.append(sum(grey_counts[:n+1])/total_pixels*max_grey)
        
    # reassemble into 2D
    
    image_2D_equal = np.array(image_1D_equal).reshape(image_2D.shape) 
    return image_2D_equal
```

*(b) Test your function on the ascent image.*

```python
plt.gray()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,12))
ax1.set_title('Original Pic')
ax1.imshow(ascent)
ax2.set_title('Histogram Equalized')
ax2.imshow(hist_equal(ascent))
plt.show()
```

***2. Convolution (Correlation) to Find a Pattern.*** *We’ve already seen how correlation can be used to find a sinc function in noise. Here we will take that same idea to find a pattern in an image.*

*(a) Load the image 400ran.png Your task is to find all the 2’s in the image. The following bit of code will grayscale the colour png image with 128 grayscale levels.*

         def rgb2gray(rgb):
             return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
         gray1 = np.floor(rgb2gray(img1)*128+0.5)

```python
ran400=mpimg.imread("./files/Lab8andWS9Images/400ran.png")
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
ran400_grey = np.floor(rgb2gray(ran400)*128+0.5)
```

```python
plt.gray()
plt.figure(figsize=(8,8))
plt.imshow(ran400_grey)
plt.show()
```

*(b) Make a template image of the 2. You can just clip out a 2 from the main image with*

         template = np.copy(img[y1:y2, x1:x2])
         template-=template.mean()
         
*where the 2D list holding the image is called img and y1:y2, x1:x2 are the pixels in the image counting from the top left. The second line scales the template to increase the contrast.*

```python
# took an educated guess where a 2 and used this plot to make sure there was nothing else
#   in the image
plt.gray()
plt.figure(figsize=(4,4))
plt.imshow(ran400_grey[0:22,0:22])
plt.show()
```

```python
# make an image according to the image location found above and scale to only include the 2
image_of_2 = np.copy(ran400_grey[6:18,12:20])
image_of_2-=image_of_2.mean()
plt.gray()
plt.figure(figsize=(4,4))
plt.imshow(image_of_2)
plt.show()
```

*(c) Correlate the template with the image with*

    signal.correlate2d(img,template,boundary=’symm’, mode=’same’).

```python
finding_2 = signal.correlate2d(ran400_grey,image_of_2,boundary='symm', mode='same')
```

*(d) Plot the 2D correlation. The 2’s should show up as bright spots.*

```python
plt.gray()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,12))
ax1.set_title('Original Pic')
ax1.imshow(ran400_grey)
ax2.set_title('2s Locations')
ax2.imshow(finding_2)
plt.show()
```

*(e) Plot a histogram of the values in the correlation. The bright spots should be the points farthest
right.*

```python
# scale the correlation data
range_finding_2 = finding_2.max()-finding_2.min()
bin_size = int(range_finding_2/128)
w,h=len(finding_2),len(finding_2) #w for width and h for height
new_finding_2 = np.zeros((w,h))
i=0
while i < w:
    j=0
    while j < h:
        new_pixel = (finding_2[i][j]+(-finding_2.min()))//bin_size
        new_finding_2[i][j] = new_pixel
        j+=1
    i+=1
```

```python
# finding the histogram of the correlation
plt.hist(finding_2.ravel(), bins=100)
plt.show()
```

*(f) One way to increase the visibility of the bright spots is to set all values of of the correlation to zero, unless the value is above some threshold. Decide on a threshold from your histogram in the previous question and zero everything else in the correlation.*

```python
# setting a threshold to only include the bright spots of the correlation. Everyting
#  below the threshold is set to zero and anything above it is increase to max
i=0
new_finding_2_dots = np.zeros((w,h))
while i < w:
    j=0
    while j < h:
        new_pixel = (finding_2[i][j]+(-finding_2.min()))//bin_size
        if new_pixel < 88:
            new_pixel = 0
        elif new_pixel >= 88:
            new_pixel = 127
        new_finding_2_dots[i][j] = new_pixel
        j+=1
    i+=1
```

*(g) Plot the 2D correlation again after the zero’ing. The image should be black with only white spots at the location of the 2’s.*

```python
plt.gray()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,12))
ax1.set_title('Original Pic')
ax1.imshow(ran400_grey)
ax2.set_title('2s Locations')
ax2.imshow(new_finding_2_dots)
plt.show()
```

***3. Dice Image.** *Image addition and subtraction can be used to combine images in different ways. Load dice_gray.png and avocado_gray.png. Devise a method to combine the images to produce the following output.*


```python
image_dice=mpimg.imread("./files/Lab8andWS9Images/dice_gray.png")
image_avacado=mpimg.imread("./files/Lab8andWS9Images/avocado_gray.png")
dice_grey = np.floor(rgb2gray(image_dice)*128+0.5)
avacado_grey = np.floor(rgb2gray(image_avacado)*128+0.5)
plt.gray()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,12))
ax1.set_title('Dice Image')
ax1.imshow(dice_grey)
ax2.set_title('Avacado Image')
ax2.imshow(avacado_grey)
plt.show()
```

```python
# make a revese silouette of the dice image
white_dice_1D= []
dice_grey_1D= dice_grey.ravel()

for n in dice_grey_1D:   # set all white to black and set everything else to white
    if n >= 127:
        new_pixel = 0
    elif n > 0:
        new_pixel = 127        
    white_dice_1D.append(new_pixel)  
white_dice_2D = np.array(white_dice_1D).reshape(dice_grey.shape) 
```

```python
plt.gray()
plt.imshow(white_dice_2D)
plt.title('Dice image cut out and reversed silouetted')
plt.show()
```

```python
# this function cuts out the shape of one image from another image

def image_removal(image1, image2):
    image1_1D = image1.ravel()
    image2_1D = image2.ravel()
    result_image = []
    result_image_1D = []
    for n1, n2 in zip(image1_1D, image2_1D):
        sub_num = n1 - n2   
        if sub_num < 0:
            sub_num = 0
        result_image.append(sub_num)
    result_image_2D = np.array(result_image).reshape(image1.shape) 
    return result_image_2D
```

```python
# make a hole for where the dice image will be added so that the images will not be
#  overlayed and only the dice image will show
avacado_minus_dice = image_removal(avacado_grey,white_dice_2D)
plt.gray()
plt.imshow(avacado_minus_dice)
plt.title('Avacado image with dice image removed')
plt.show()
```

```python
# this function adds two images together

def image_addition(image1, image2):
    image1_1D = image1.ravel()
    image2_1D = image2.ravel()
    result_image = []
    result_image_1D = []
    for n1, n2 in zip(image1_1D, image2_1D):
        add_num = n1 + n2
        if add_num >127:
            add_num = 127
        result_image.append(add_num)
    result_image_2D = np.array(result_image).reshape(image1.shape) 
    return result_image_2D
```

```python
# make the background of the dice image black so that when it is added to another image 
#  the white background of this image will not interfere with the other the other image.
black_back_dice_1D = []
for n in dice_grey_1D:
    new_pixel = n
    if n >= 127:
        new_pixel = 0       
    black_back_dice_1D.append(new_pixel)  
black_back_dice_2D = np.array(black_back_dice_1D).reshape(dice_grey.shape) 
```

```python
plt.gray()
plt.imshow(black_back_dice_2D)
plt.title('Dice image with background removed')
plt.show()
```

```python
# add the image of avacado background and the image of the dice with background removed to 
#   produce the target image. 
dice_avacado = image_addition(avacado_minus_dice,black_back_dice_2D)

plt.gray()
plt.figure(figsize=(10,10))
plt.imshow(dice_avacado)
plt.title('Dice image with Avacado Background')
plt.show()
```
