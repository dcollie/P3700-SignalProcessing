Discrete Time Fourier Transform (DTFT). The information provided by the the DTFT is called the filter’s frequency response. The frequency response consists of the magnitude response and the phase response.


The definition for the DTFT of a signal $x[n]$ is

$$ X(\Omega) = \sum_{n=-\infty}^{\infty}x[n]e^{-jn\Omega} $$

or

$$ X(\Omega) = \sum_{n=-\infty}^{\infty}x[n](cos(n\Omega)-jsin(n\Omega)) $$

Reminder that the z transform is

$$ X(\Omega) = \sum_{n=0}^{\infty}x[n]z^{-n} $$
