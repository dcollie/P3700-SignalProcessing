```python
def f_response(omega):
    return 0.2*(1+np.exp(-1j*omega)+np.exp(-2j*omega)+np.exp(-3j*omega)+np.exp(-4j*omega))

df_space = np.arange(0, np.pi/2, np.pi*1/64)
```
