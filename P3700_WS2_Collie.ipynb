```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,9)
plt.axis([0,8,0,8])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantized Level")
plt.plot(x)
plt.step(x,x-1)
plt.show()
```

TASK 1a

```python

x1 = np.arange(0,9,.1)      # 0 - 9 with 10 samples per number
x2 = np.floor(x1)           # 0 - 9 quantized signal
q_error = x2-x1             # error between the analogue signal and quantized signal

plt.axis([0,8,-1,0])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantized Error")
plt.plot(x1,q_error)
plt.show()
```

TASK 1b

```python
# maximum error
max_error = abs(min(q_error))                 # assign abs value of min error to max error
if (abs(min(q_error)) < abs(max(q_error))):   #   but if max error is bigger than abs min error
    max_error = abs(max(q_error))             #   then max error is the max error
print ("Maximum quantization error: ",max_error)
```

TASK 1c

```python
x2_half = x2+.5                 # shift by .5
q_error_half = x2_half-x1       # error between the code

plt.axis([0,8,-.75,.75])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantized Error")
plt.plot(x1,q_error_half)
plt.show()
```

TASK 1d

TASK 2

```python
A=2  # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=100.0*f
Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=50  # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve


plt.figure(figsize=(12,4))
plt.axis([0,2,-2,2]) # were not plotting all cycles so we can see whats going on
plt.plot(t,x)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("sin Curve")
plt.show()
```

TASK 2a

```python
# overlaying a stem plot at sampling frequency of 30TASK 2dHz
fs_30 = 30
Ts_30 = 1/fs_30
t_30 = np.arange(0,cycles*T,Ts_30)
x_30 = A*np.sin(2*np.pi*f*t_30)

plt.figure(figsize=(12,4))
plt.axis([0,2,-2,2]) # were not plotting all cycles so we can see whats going on
plt.plot(t,x)
plt.stem(t_30,x_30)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Sin Curve with Stem Plot at 30Hz")
plt.show()
```
