```python
import numpy as np
import matplotlib.pyplot as plt
```

TASK 2

```python
avg_wind_speed = np.loadtxt("weatherstats_vancouver_daily.csv", delimiter=",", usecols=(17), skiprows=(1))

mean_aws = np.mean(avg_wind_speed)
std_aws = np.std(avg_wind_speed)
var_aws = np.var(avg_wind_speed)
stn_aws = mean_aws/std_aws
print ("Mean of the Average Daily Wind Speeds: %.3f km/h."%(mean_aws))
print ("STD of the Average Daily Wind Speeds: %.3f km/h" % (std_aws))
print ("Variance of the Average Daily Wind Speeds: %.3f km/h" %(var_aws))
print ("Signal to Noise of the Average Daily Wind Speeds: %.3f km/h" %(stn_aws))
```

TASK 3

```python
plt.hist(avg_wind_speed, bins='auto')
plt.xlabel('Average Wind Speed (km/h)')
plt.ylabel('Days')
plt.title('YVR Average Wind Speed Data')
plt.show()
```

TASK 4

```python
n, bins, patches=plt.hist(avg_wind_speed, 50)
plt.setp(patches[15], 'facecolor', 'r')
plt.text(20, 400, r'$\bar{x}=$%.3f   $\sigma=$%.3f' % (mean_aws, std_aws))
plt.axis([0, 50, 0,1300])
plt.xlabel('Average Wind Speed (km/h)')
plt.ylabel('Days')
plt.title('YVR Average Wind Speed Data')
plt.grid(True)
plt.show()
```

```python
rand_A = np.random.randint(1, 100, 10000)
rand_B = np.random.randint(1, 100, 10000)
sum_AB = rand_A+rand_B
mean_AB = np.mean(sum_AB)
std_AB = np.std(sum_AB)
```

```python
plt.hist(sum_AB, bins='auto')
plt.text(150, 500, r'$\bar{x}=$%.3f' % (mean_AB)) 
plt.text(150, 450, r'$\sigma=$%.3f' % (std_AB))
plt.xlabel('Sum of Random Numbers')
plt.ylabel('# of Occurances')
plt.title('Sum of 2 Random Numbers')
plt.show()
```

```python
normalsum = sum_AB
for i in range(10):
    normalsum = normalsum + np.random.randint(1, 100, 10000)
mean_normal = np.mean(normalsum)
std_normal = np.std(normalsum)
```

```python
plt.hist(normalsum, bins='auto')
plt.text(800, 400, r'$\bar{x}=$%.3f' % (mean_normal)) 
plt.text(800, 350, r'$\sigma=$%.3f' % (std_normal))
plt.xlabel('Sum of Random Numbers')
plt.ylabel('# of Occurances')
plt.title('Sum of 12 Random Numbers')
plt.show()
```
